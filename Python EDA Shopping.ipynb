```python
#EDA
ax=sns.countplot(x="Gender",data= data)
```

```python

ax=sns.countplot(x="Gender",data= data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
data.groupby(["Amount","Gender"]).sum()
```

```python
data.groupby(["Amount","Gender"]).sum().sort_values(by="Amount",ascending=False)
```

```python
data.groupby(["Amount"],as_index=False)["Gender"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales=data.groupby(["Amount","Gender"]).sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data= sales)
```

```python
sns.countplot(data=data,x="Age Group",hue="Gender")
```

```python
ax=sns.countplot(data=data,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=data.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending = False)
sns.barplot(x="Age Group",y="Amount",data=sales_age,palette="viridis")
```

```python
orders_states=data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(19,5)})
sns.barplot(x="State",data=orders_states,y="Orders",palette="magma")
#to incease area of bar plot
sns.set(rc={"figure.figsize":(15,5)})
```

```python
#total amount/sales from Top 10 states
sales_states=data.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set({"figure.figsize":(15,5)})
sns.barplot(x="State",data=sales_states,y="Amount",palette="coolwarm")
```

```python
#marital status
ax=sns.countplot(x="Marital_Status",data= data,palette="dark")
sns.set({"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marriage=data.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set({"figure.figsize":(6,5)})
sns.barplot(x="Marital_Status",data=sales_marriage,y="Amount",hue="Gender",palette="RdBu")
```

```python
ax=sns.countplot(x="Occupation",data= data,palette="Blues")
sns.set({"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=data.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set({"figure.figsize":(16,5)})
sns.barplot(x="Occupation",data=sales_occupation,y="Amount",palette="cubehelix")
```

```python
# product category
ax=sns.countplot(x="Product_Category",data= data,palette="colorblind")
sns.set({"figure.figsize":(17,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_productcategory=data.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set({"figure.figsize":(16,5)})
sns.barplot(x="Product_Category",data=sales_productcategory,y="Amount",palette="pastel")
```

```python
sales_productid=data.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set({"figure.figsize":(16,5)})
sns.barplot(x="Product_ID",data=sales_productid,y="Orders",palette="muted")
```
